```python
#labaries(external tool)
import pandas as pd #for handling and analyzing
import numpy as np #for numerical operstions
import matplotlib.pyplot as plt #for visulization like graph, chart etc.
import seaborn as sns # seabron bulids on matplotlib to create asenthetically better statistical charts.

#create and collect ur data
data ={'Fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Banana', 'Orange', 'Apple', 'Banana'], 'Price_per_kg': [120, 60, 80, 125, 55, 75, 130, 65],
 'Quantity_Sold': [30, 50, 40, 35, 60, 42, 25, 551], 'Day': ['Mon', 'Mon', 'Mon', 'Tue', 'Tue', 'Tue', 'Wed', 'Wed']}

df = pd.DataFrame(data)

#cleaning the data (step 2)
print(df.isnull().sum()) # to find missing values
#fix the dataset
df['Price_per_kg'] = df['Price_per_kg'].astype(float) #astype as num with decimals(float)
df['Quantity_Sold'] = df['Quantity_Sold'].astype(int) #astype as num with decimals(float)

#EDA = to understand more about data(they have three steps)
#check first few rows
print(df.head()) #.head it will show the first 5 rows
print(df.tail()) #.tail it will show the last 5 rows

#Basic statistics min, max, min etc (step4)

print(df.describe())

#data patterns (step5)
sales_by_day = df.groupby('Day')['Quantity_Sold'].sum()
print(sales_by_day)

#data visualization (story-driven EDA)
#simple bar chart for fruit sales
sns.barplot(x='Fruit', y='Quantity_Sold', data=df)
plt.title('Fruit Sales')
plt.show()

#data storytelling & bias checking(step 6)
#bad chart
sns.barplot(x='Day', y='Quantity_Sold', data=df)
plt.ylim(20,65)
plt.title('Fruit Sales')
plt.show()

#good chart
sns.barplot(x='Day', y='Quantity_Sold', data=df)
plt.ylim(0,65)
plt.title('Fruit Sales')
plt.show()

#Desinging the Charts (step 7)
sns.barplot(x='Day', y='Quantity_Sold', data=df, hue='Day', palette= 'colorblind', legend=False)
plt.ylim(0,65)
plt.title('Fruit Sales')
plt.show()
```
